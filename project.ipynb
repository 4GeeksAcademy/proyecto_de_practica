#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps, Marker,AwesomeIcon

# Map centered on (60 degrees latitude and -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (40.3165, -3.8667), zoom = 11, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenTopoMap)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
#Creating a dictionary to get latitudes and longitudes 
dic21 = cinturonsur.set_index('Unnamed: 0')[['latitude', 'longitude', 'level5']].to_dict(orient='index')


#Assigning colors to cities to implement AwesomeIcon
colores_ciudades = {
    'Alcorcón': 'orange',
    'Getafe': 'darkgreen',
    'Fuenlabrada': 'red',
    'Leganés': 'beige'
}

#Latitude/longitude converting to proper format ,/.
for nim, valores in dic21.items():
    latitude = float(valores['latitude'].replace(',', '.'))
    longitude = float(valores['longitude'].replace(',', '.'))
    level5 = valores['level5']

#Setting AwesomeIcon parameters
    icono = AwesomeIcon(name='home',  marker_color=colores_ciudades.get(level5, 'blue')) 
    
#Assigning markers by city.
    if level5 == 'Alcorcón':
        marcador = Marker(location=(latitude, longitude), icon=icono , draggable=True)
        map.add_layer(marcador)
    if level5 == 'Getafe':
        marcador = Marker(location=(latitude, longitude),icon=icono , draggable=True)
        map.add_layer(marcador)
    if level5 == 'Fuenlabrada':
        marcador = Marker(location=(latitude, longitude),icon=icono , draggable=True)
        map.add_layer(marcador)
    if level5 == 'Leganés':
        marcador = Marker(location=(latitude, longitude),icon=icono , draggable=True)
        map.add_layer(marcador)

map


#¿Por qué hay un inmueble en Getafe que tiene las coordenadas en la costa oeste de África? Me quedo con la duda. 
```
